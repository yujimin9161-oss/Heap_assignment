```python
import csv
import heapq

heap = []

with open('birthday.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # 헤더 제거

    for row in reader:
        # 빈 줄 방지
        if len(row) == 0:
            continue

        name = row[0]

        # 생일 없는 경우 제외
        if row[1] == '' or row[2] == '' or row[3] == '':
            continue

        year = row[1]
        month = row[2].zfill(2)
        day = row[3].zfill(2)

        birthday = int(year + month + day)

        # 최대 힙 만들기 위해 음수 사용
        heapq.heappush(heap, (-birthday, name))

print("생일이 느린 순서 10명:")

for _ in range(min(10, len(heap))):
    b, name = heapq.heappop(heap)
    print(name, -b)
```
